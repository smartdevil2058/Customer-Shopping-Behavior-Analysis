```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Category')['Review Rating'].transform('median'))
```

```python
labels = ['young adult', 'Adult', 'Middle Aged', 'Senior']
df['Age Group'] = pd.qcut(df['Age'], q = 4, labels=labels)
```

```python
df[['Age', 'Age Group']]
```

```python
Frequency = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['Frequency of Purchases(numbers)'] = df['Frequency of Purchases'].map(Frequency)
```

```python
df[['Frequency of Purchases', 'Frequency of Purchases(numbers)']]
```
